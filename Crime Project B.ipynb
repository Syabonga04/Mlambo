# <b><span style='color:#fcc36d'>Step 4.3 |</span><span style='color:#ff6200'> Outlier Treatment</span></b>
⬆️ [Tabel of Contents](#contents_tabel)

<div style="border-radius:10px; padding: 15px; background-color: #ffeacc; font-size:120%; text-align:left">
    
Next, I am aimimg at identifying potential outliers in a dataset and treat outliers in the dataset.

```python
print("\nBasic statistics for numeric columns:")
if crime_data.select_dtypes(include='number').shape[1] > 0: 

        print("\nDetecting outliers:")
        for col in crime_data.select_dtypes(include='number').columns:
            Q1 = crime_data[col].quantile(0.25)
            Q3 = crime_data[col].quantile(0.75)
            IQR = Q3 - Q1
            
            # Calculate outlier bounds
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            # Detect outliers
            outliers = crime_data[(crime_data[col] < lower_bound) | (crime_data[col] > upper_bound)]
            
            print(f"\nColumn: {col}")
            print(f"Lower Bound: {lower_bound}")
            print(f"Upper Bound: {upper_bound}")
            print(f"Number of Outliers: {outliers.shape[0]}")
            
            if outliers.shape[0] > 0:
                # Optionally display a sample of outliers
                print("\nSample of outliers:")
                print(tabulate(outliers.head(), headers='keys', tablefmt='grid'))

                # Plot boxplot for the column
                plt.figure(figsize=(10, 6))
                sns.boxplot(data=crime_data, y=col)
                plt.title(f'Boxplot for {col}')
                plt.show()
else:
        print("No numeric columns found.")
```

- Convert 'Crime Count' and 'Province Population' to numeric.
-  The crime rate can be calculated using the formula:Crime Rate=(Crime Count/Province Population)*100000.This formula gives the crime rate per 100,000 people,

```python
def convert_to_numeric_preserving_zeros(column):
    return pd.to_numeric(column, errors='coerce').fillna(0)


crime_data['Crime Count'] = convert_to_numeric_preserving_zeros(crime_data['Crime Count'])
crime_data['Province Population'] = convert_to_numeric_preserving_zeros(crime_data['Province Population'])
```

```python
crime_data['Crime Rate'] = (crime_data['Crime Count'] / crime_data['Province Population']) * 100000
```

```python
crime_data['Start Year Date'] = pd.to_numeric(crime_data['Start Year Date'], errors='coerce')
```

```python
average_crime_rates = crime_data.groupby('Province')['Crime Count'].mean()
total_crime_counts = crime_data['Crime Count'].sum()
print(f"Total Crime Counts: {total_crime_counts}")
print("Average Crime Rates per Province:")
print(average_crime_rates)
```

```python
early_period = crime_data[(crime_data['Start Year Date'] >= 2013) & (crime_data['Start Year Date'] <= 2015)]
recent_period = crime_data[(crime_data['End Year Date'] >= 2021) & (crime_data['End Year Date'] <= 2023)]
early_trends = early_period.groupby('Start Year Date')['Crime Count'].sum().reset_index()
recent_trends = recent_period.groupby('End Year Date')['Crime Count'].sum().reset_index()

comparison = pd.DataFrame({
    'Year': early_trends['Start Year Date'].astype(str).tolist() + recent_trends['End Year Date'].astype(str).tolist(),
    'Total Crime Count': early_trends['Crime Count'].tolist() + recent_trends['Crime Count'].tolist()
})

plt.figure(figsize=(10, 6))
plt.bar(comparison['Year'], comparison['Total Crime Count'], color=['orange' if year <= '2015' else 'green' for year in comparison['Year']])
plt.title('Comparison of Early (2013-2015) and Recent (2021-2023) Crime Trends')
plt.xlabel('Year')
plt.ylabel('Total Crime Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
for col in crime_data.select_dtypes(include=[np.datetime64]).columns:
    crime_data[col] = crime_data[col].astype('int64') // 10**9
```

```python
crime_data = crime_data.drop(columns=['Crime Rate'])
crime_data
```

```python

X = crime_data.drop('Crime Count', axis=1)
y_classification = (crime_data['Crime Count'] > 150).astype(int)  # Example binary target for Logistic Regression
y_regression = crime_data['Crime Count']
```

```python
categorical_cols = ['Type of Crime', 'Province','Crime Category']
numeric_cols = ['Province Population']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_cols),
        ('cat', OneHotEncoder(), categorical_cols)
    ])
```

```python

X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X, y_classification, test_size=0.2, random_state=42)

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_regression, test_size=0.2, random_state=42)

classification_pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

X_train_class_transformed = classification_pipeline.fit_transform(X_train_class)
X_test_class_transformed = classification_pipeline.transform(X_test_class)
```

```python

rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train_class_transformed, y_train_class)
```

```python
y_pred_rf = rf_classifier.predict(X_test_class_transformed)
rf_accuracy =accuracy_score(y_test_class, y_pred_rf)
```

```python

feature_names = numeric_cols + list(classification_pipeline.named_steps['preprocessor'].transformers_[1][1].get_feature_names_out(categorical_cols))

rf_importances = rf_classifier.feature_importances_

rf_indices = np.argsort(rf_importances)[::-1]
sorted_rf_features = [feature_names[i] for i in rf_indices]
sorted_rf_importances = rf_importances[rf_indices]

plt.figure(figsize=(12, 12))
plt.barh(sorted_rf_features, sorted_rf_importances, color='skyblue', edgecolor='black', height=0.4)
plt.title("Feature Importance for Random Forest Classifier", fontsize=20)
plt.xlabel("Importance", fontsize=16)
plt.ylabel("Features", fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.xlim(0, max(sorted_rf_importances) * 0.9)  
for index, value in enumerate(sorted_rf_importances):
    plt.text(value, index, f"{value:.2f}", va='center', fontsize=12)

plt.tight_layout()
plt.show()
```
